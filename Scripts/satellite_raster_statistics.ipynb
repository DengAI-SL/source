```python
tiff_directory = "/home/gayan/Desktop/dengai/Data/Satellite/raster/satellite_2012-01-01_2012_06_01"
```

```python
satellite_stat = []
for filename in os.listdir(tiff_directory):
    if filename.endswith(".tif"): 
        file_abs = os.path.join(tiff_directory, filename)
        region_id = filename.split('_')[-1].split('.')[0]
        ds = gdal.Open(file_abs)
        raster_array = np.array(ds.GetRasterBand(1).ReadAsArray())
#         raster_interp = np.interp(raster_array, (np.nanmin(raster_array), np.nanmax(raster_array)), (0, +1))
        satellite_stat.append({'REGION_ID':region_id, 'NDVI_MEAN':np.nanmean(raster_array),'NDVI_STD':np.nanstd(raster_array),'NDVI_MIN':np.nanmin(raster_array),'NDVI_MAX':np.nanmax(raster_array)})
        continue
    else:
        continue
satellite_stat_df = pd.DataFrame(satellite_stat)
```
